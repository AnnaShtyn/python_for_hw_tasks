2. **Опціонально (по 0.5 бала за кожну доцільну ознаку):** Додайте нові ознаки, обчислені на основі наявних даних, які б на ваш погляд були корисними для моделі

```python
# 1. Не будемо кодувати стовпці 'Floor' та 'Area Locality' оскільки там забагато
#    значень, і модель не зможе знайти патерни.

area_type_dummies = pd.get_dummies(df_clean['Area Type'], prefix='area_type').astype(int)
city_dummies = pd.get_dummies(df_clean['City'], prefix='city').astype(int)
furnishing_dummies = pd.get_dummies(df_clean['Furnishing Status'], prefix='furnishing').astype(int)
tenant_dummies = pd.get_dummies(df_clean['Tenant Preferred'], prefix='tenant').astype(int)
contact_dummies = pd.get_dummies(df_clean['Point of Contact'], prefix='contact').astype(int)
```

```python
# 3. додамо нові закодовані колонки до датафрейму

df_clean_new = pd.concat([df_clean, area_type_dummies, city_dummies, furnishing_dummies, tenant_dummies, contact_dummies], axis=1)

#  та видалимо зайві (текстові), і перевіримо остаточні колонки нового датафрейму
df_clean_new = df_clean_new.drop(['Floor', 'Area Locality', 'Posted On', 'Area Type', 'City', 'Furnishing Status', 'Tenant Preferred', 'Point of Contact'], axis=1)
df_clean_new.info()
```

```python
# 4. Отже, оскільки в нас оригінальні числові ознаки досить суттєві (згідно з кор.)
#   матрицею), ми беремо їх та закодовані категоріальні ознаки (тобто всі ознаки з
#   df_clean_new окрім цільової 'Rent')

x = df_clean_new.drop(['Rent'], axis=1)
y = df_clean_new['Rent']
print(x.shape, y.shape)
```

```python
# 5.
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled_df = pd.DataFrame(x_scaled, columns=x.columns, index=x.index)
x_scaled_df.head()
```

## Завдання 5: Розділення даних та навчання моделі (3 бали)

**Що потрібно зробити:**
1. Розділіть дані на навчальну (80%) та тестову (20%) вибірки.
2. Створіть модель лінійної регресії.
3. Навчіть модель на навчальних даних.
4. Виведіть усі коефіцієнти моделі (ваги) та напишіть, які 2 ознаки найбільше впливають на прогноз.
5. Зробіть прогнози на тренувальній та тестовій вибірках.

```python
# 1.
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x_scaled_df, y, test_size=0.2, random_state=10
)
```

```python
# 2,3.
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# 4.
for feature, weight in zip(model.feature_names_in_, model.coef_):
    print(f'{feature}: {weight:.2f}')

print(f'Незалежний параметр (intercept): {model.intercept_:.2f}')

coef_imp = pd.Series(model.coef_, index=x.columns).sort_values(key=np.abs, ascending=False).round(2)
print(f'\nТоп-2 найвагоміших коефіцієнтів ознак\n{coef_imp.head(2)}')
```

```python
# 5.
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)

comp = pd.DataFrame({
    'Реальна оренд.плата' : y_test.values[:10],
    'Прогнозована оренд.плата' : y_test_pred[:10].round(0),
    'Похибка' : y_test.values[:10] - y_test_pred[:10].round(0)
})
print(f'Приклад прогнозування на тестовій вибірці:\n{comp}')
```

```python
# 1,2.
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

MAE = mean_absolute_error(y_test, y_test_pred)
MSE = mean_squared_error(y_test, y_test_pred)
RMSE = np.sqrt(MSE)
R2 = r2_score(y_test, y_test_pred)
print(f'Показники якості моделі на ТЕСТОВІЙ вибірці:\nMAE: {MAE:.2f}\nMSE: {MSE:.2f}\nRMSE: {RMSE:.2f}\nR²:{R2:.2f}')

MAE_tr = mean_absolute_error(y_train, y_train_pred)
MSE_tr = mean_squared_error(y_train, y_train_pred)
RMSE_tr = np.sqrt(MSE_tr)
R2_tr = r2_score(y_train, y_train_pred)
print(f'\nПоказники якості моделі на ТРЕНУВАЛЬНІЙ вибірці:\nMAE: {MAE_tr:.2f}\nMSE: {MSE_tr:.2f}\nRMSE: {RMSE_tr:.2f}\nR²:{R2_tr:.2f}')
```

```python
# 1,2.
resid = y_test - y_test_pred

fig = px.histogram(
    x=resid,
    nbins=50,
    title='Розподіл помилок прогнозування орендної плати',
    labels={'x' : 'Помилки (реал-прогноз)', 'count' : 'Кількість'}
)
fig.add_vline(x=0, line_dash='dash', line_color='red', annotation_text='Ідеальний прогноз')
fig.show()
```

```python
# 3.
fig= px.scatter(
    x=y_test_pred,
    y=resid,
    title='Залежність помилок від прогнозованих значень орендної плати',
    labels={'x' : 'Прогнозована ор.плата', 'y' : 'Помилка'},
    opacity=0.5
)
fig.add_hline(y=0, line_dash='dash', line_color='black', annotation_text='Відсутність помилки')
fig.show()

# Так, з ростом прогнозованого значення ор. плати похибка збільшується в обидві сторони.
```

```python
# 4.
errors_df = pd.DataFrame({
'real' : y_test.values,
'pred' : y_test_pred.round(2),
'error' : np.abs(resid).round(2)
})

top5_errors = errors_df.nlargest(5, 'error')
print(f'Топ-5 прогнозованих орендних плат з найбільшими помилками\n {top5_errors}')
```

```python
# 5. Створимо таблицю абсолютних значень топ5 найбільш помилкових прогнозів, де
#   наочно бачимо 'вагу' кожної ознаки на передбачення орендної плати.

df_top5_errors_fin = x_scaled_df.loc[top5_errors.index]
df_weights = coef_imp.to_frame().T[df_top5_errors_fin.columns]

df_top5_errors_fin = df_top5_errors_fin.mul(coef_imp, axis=1).round(2).transpose()
top5_errors_fin_tot = df_top5_errors_fin.sum(axis=0)
df_top5_errors_fin.loc['Total+intercept'] = (top5_errors_fin_tot + 19399.42).round(0)

display(df_top5_errors_fin)

# Отож, для позиції 3912 прогнозоване значення набагато нижче ймовірно тому, що
#   реальна орендна плата в оголошенні була безпідставно завищена.
#  Для решти 4 оголошень прогнозована ціна значно вища через незалежний параметр
# (19499), к-сть кімнат, розмір, к-сть ванних кімнат,контактної особи та локації
#  в Мумбаї (в кожному випадку різний набір факторів).
```
